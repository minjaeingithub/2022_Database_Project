```python
import matplotlib.pyplot as plt
```

```python
# Plot when changing journal mode
# metric : milliseconds
x = ["off", "delete", "wal"]
y = [215, 1232, 136]
for i, v in enumerate(x):
    plt.text(v, y[i], y[i])
plt.bar(x, y, width = 0.5)
# plt.xlabel("journal mode")
plt.ylabel("real time in milliseconds")
plt.title("Changing Journal mode")

# plt.show()
plt.savefig("sqlite/journal_mode.png")
```

```python
# Plot when changing page cache size
# metric : milliseconds
x = ["100", "500", "1000", "2000"]
y = [1128, 914, 1066, 780]

for i, v in enumerate(x):
    plt.text(v, y[i], y[i])
    
plt.bar(x, y, width = 0.3)
# plt.xlabel("journal mode")
plt.xticks(x);
plt.ylabel("real time in milliseconds")
plt.title("Changing page cache size")

# plt.show()
plt.savefig("sqlite/page_cache_size.png")
```
