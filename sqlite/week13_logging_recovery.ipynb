```python
import matplotlib.pyplot as plt
```

```python
x = ["DELIVERY", "NEW_ORDER", "ORDER_STATUS", "PAYMENT", "STOCK_LEVEL"]
delete = [730.02, 519.83, 3503.84, 350.02, 224.68]
wal = [330.52, 42.80, 890.78, 63.06, 52.63]

for i, v in enumerate(x):
    plt.text(v, delete[i], delete[i])
    
for i, v in enumerate(x):
    plt.text(v, wal[i], wal[i])
plt.ylabel("txn/s")
plt.title("Throughput per journal mode")

plt.plot(x, delete, label="DELETE")
plt.plot(x, wal, label="WAL")
plt.legend()


plt.savefig("./tps_per_transtype");
```
